```python
import numpy as np
```

```python
from sklearn.base import BaseEstimator

class Tree(BaseEstimator):
    def __init__(self, max_depth=3, min_data_in_leaf=3):
        self.max_depth = max_depth
        self.min_data_in_leaf = min_data_in_leaf
    
        self.tree = {(): None}
        self.cur_depth = 0
    
    @staticmethod
    def get_entropy(y):
        values, counts = np.unique(y, return_counts=True)
        if len(values) == 1:
            return 0
        p = counts / y.shape[0]
        return -(p * np.log(p)).sum()
    
    def get_split(self, X, y):
        n, m = X.shape
        E_min = np.inf
        feat_best, val_best = 0, X[0,0]
        mask_best = np.ones(X.shape[0], dtype=np.bool)
        for j in range(m):
            for t in np.unique(X[:, j]):
                mask = X[:, j] <= t
                if np.unique(mask, return_counts=True)[1].min() < 3:
                    continue
                mask_mean = mask.mean()
                E = mask_mean * self.get_entropy(y[mask]) + (1 - mask_mean) * self.get_entropy(y[~mask])
                if E < E_min:
                    E_min, feat_best, val_best, mask_best = E, j, t, mask
        return feat_best, val_best, mask_best

    def traverse(self, X, y, path=()):
        if self.cur_depth < self.max_depth:
            col, t, mask = self.get_split(X, y)
            if len(np.unique(mask)) == 1:
                return
            del self.tree[path]
            self.cur_depth += 1
            for is_greater in range(2):
                mask_leaf = ((mask + is_greater) % 2).astype(np.bool)
                X_leaf, y_leaf = X[mask_leaf], y[mask_leaf]
                path_leaf = tuple(x for x in path if x[:2] != (col, is_greater)) + ((col, is_greater, t),)
                self.tree[path_leaf] = y_leaf.mean()
                self.traverse(X_leaf, y_leaf, path_leaf)
    
#     def simplify_paths(self):
#         tree_new = {}
#         for path, v in self.tree.items():
#             d = {}
#             for j, is_greater, t in path:
#                 if (d.get((j, is_greater), -np.inf) < t and is_greater == 0) or (d.get((j, is_greater), np.inf) > t and is_greater == 1):
#                     d[(j, is_greater)] = t,
#             tree_new[tuple((k+v for k,v in d.items()))] = v
#         self.tree = tree_new
        
    def fit(self, X, y):
        self.traverse(X, y)
        
    def predict(self, X):
        y_pred = np.zeros(X.shape[0])
        for path, v in self.tree.items():
            mask = np.ones(X.shape[0])
            for j, is_greater, t in path:
                mask_j = X[:, j] <= t
                mask *= (mask_j + is_greater) % 2
            y_pred[mask.astype(np.bool)] = v
        return y_pred
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import roc_auc_score, make_scorer

n_samples = 1000
X, y = make_classification(n_features=2, n_informative=2, n_redundant=0, n_samples=n_samples, class_sep=.5)
plt.scatter(X[:,0], X[:,1], c=y, alpha=0.6);
```

```python
%%time
tree = Tree(max_depth=10)
tree.fit(X, y)
```

```python
y_pred = tree.predict(X)
```

```python
tree = Tree(max_depth=10)
scorer = make_scorer(roc_auc_score)
y_pred = cross_val_predict(tree, X, y)
scores = cross_val_score(tree, X, y, scoring=scorer, cv=10, n_jobs=-1)
scores.mean(), scores.std()
```
